```python
def main_calculator(): #ממשק משתמש
    while True:
        print("\n--- מחשבון המרת בסיסים (Hex <-> Decimal) ---")
        print("1. המרה מעשרוני לבסיס הקסדצימלי (בסיס 16)")
        print("2. המרה מבסיס הקסדצימלי (בסיס 16) לבסיס עשרוני")
        print("3. יציאה מהתוכנית")
        
        choice = input("בחר אפשרות (1-3): ").strip()
        
        if choice == '3':
            print("תודה שהשתמשת במחשבון. להתראות!")
            break
            
        if choice not in ['1', '2']:
            print("\nבחירה לא תקינה, נסה שנית.") 
            continue
            
        # לולאה פנימית להזנת קלט עד שיהיה תקין
        while True:
            value = input("הכנס ערך להמרה: ")
            
            if choice == '1':
                result = manual_decimal_to_hex(value)
                # בדיקה אם חזרה שגיאה מהפונקציה
                if isinstance(result, str) and ("שגיאה" in result or "Error" in result):
                    print(f"{result}. נסה שנית.")
                else:
                    print(f"תוצאה בבסיס הקסדצימלי (בסיס 16): {result}")
                    break # קלט תקין - יוצאים מהלולאה הפנימית חזרה לתפריט
            
            elif choice == '2':
                result = manual_hex_to_decimal(value)
                # בדיקה אם חזרה שגיאה מהפונקציה
                if isinstance(result, str) and ("שגיאה" in result or "Error" in result):
                    print(f"{result}. נסה שנית.")
                else:
                    print(f"תוצאה בבסיס עשרוני: {result}")
                    break # קלט תקין - יוצאים מהלולאה הפנימית חזרה לתפריט

# חזרה לתחילת התפריטן
main_calculator()


#המרה מבסיס עשרוני לבסיס 16
def manual_decimal_to_hex(dec_input):
    try:
        n = float(dec_input)
    except ValueError:
        return "שגיאה: קלט לא תקין"

    # טיפול בסימן שלילי
    is_negative = n < 0
    n = abs(n)
    
    # הפרדה לחלק שלם וחלק שבר
    integer_part = int(n)
    fractional_part = n - integer_part
    
    hex_chars = "0123456789ABCDEF"
    
    # המרת החלק השלם (בשיטת החילוק)
    int_result = ""
    temp_int = integer_part
    if temp_int == 0:
        int_result = "0"
    else:
        while temp_int > 0:
            int_result = hex_chars[temp_int % 16] + int_result
            temp_int //= 16
            
    # המרת החלק השברי (בשיטת הכפל)
    frac_result = ""
    # נגביל ל-6 ספרות אחרי הנקודה כדי למנוע לולאה אינסופית בשברים מחזוריים
    precision = 6 
    
    while fractional_part > 0 and len(frac_result) < precision:
        fractional_part *= 16
        digit = int(fractional_part) # לוקחים את החלק השלם שנוצר
        frac_result += hex_chars[digit]
        fractional_part -= digit # משאירים רק את השארית השברית להכפלה הבאה
        
    # חיבור התוצאה
    final_hex = int_result
    if frac_result:
        final_hex += "." + frac_result
        
    return "-" + final_hex if is_negative else final_hex





#המרה מבסיס 16 לבסיס עשרוני
def manual_hex_to_decimal(hex_input):

    # הכנת הקלט - ניקוי רווחים והפיכה לאותיות גדולות
    hex_str = str(hex_input).strip().upper()
    if not hex_str:
        return "שגיאה: קלט ריק"

    # טיפול בסימן שלילי
    is_negative = False
    if hex_str.startswith('-'):
        is_negative = True
        hex_str = hex_str[1:]
    
    # בדיקת תקינות התווים (can write numbers 0-9, letter A-F, only one dot)
    hex_chars = "0123456789ABCDEF"
    if hex_str.count('.') > 1:
        return "שגיאה: יותר מנקודה עשרונית אחת"
    
    for char in hex_str:
        if char not in hex_chars and char != '.':
            return f"שגיאה: התו '{char}' לא חוקי בבסיס 16"

    # הפרדה לחלק שלם ושבר
    if '.' in hex_str:
        integer_part, fractional_part = hex_str.split('.')
    else:
        integer_part, fractional_part = hex_str, ""

    decimal_sum = 0.0

    # חישוב החלק השלם: ספרה כפול 16 בחזקת המיקום (מימין לשמאל)
    for i, char in enumerate(reversed(integer_part)):
        digit_value = hex_chars.find(char)
        decimal_sum += digit_value * (16 ** i)

    # חישוב החלק השברי: ספרה כפול 16 בחזקת מיקום שלילית (משמאל לימין)
    for i, char in enumerate(fractional_part):
        digit_value = hex_chars.find(char)
        decimal_sum += digit_value * (16 ** -(i + 1))

    # החזרת תוצאה (הפיכה לסימן הנכון)
    final_result = -decimal_sum if is_negative else decimal_sum
    
    # מבחינת אסתטיקה, אם המספר שלם נחזיר אותו כאינטג'ר, אם לא אז כפלואוט
    if final_result == int(final_result):
        return int(final_result)
    
    return final_result
```
