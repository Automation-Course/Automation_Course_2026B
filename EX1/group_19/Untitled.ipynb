```python
def decimal_to_hex(decimal_number):
    if not isinstance(decimal_number, (int, float)):
        raise ValueError("Input must be a number.")

    hex_digits = "0123456789ABCDEF"

    # Sign and magnitude
    sign = ""
    if decimal_number < 0:
        sign = "-"
        decimal_number = -decimal_number

    # Split into integer and fractional parts
    integer_part = int(decimal_number)
    fractional_part = decimal_number - integer_part

    # Convert integer part
    if integer_part == 0:
        integer_hex = "0"
    else:
        integer_hex = ""
        while integer_part > 0:
            remainder = integer_part % 16
            integer_hex = hex_digits[remainder] + integer_hex
            integer_part //= 16

    # Convert fractional part
    fractional_hex = ""
    precision = 5

    while fractional_part > 0 and len(fractional_hex) < precision:
        fractional_part *= 16
        digit = int(fractional_part)
        fractional_hex += hex_digits[digit]
        fractional_part -= digit

    if fractional_hex == "":
        return sign + integer_hex
    else:
        return sign + integer_hex + "." + fractional_hex


def hex_char_to_value(ch):
    if '0' <= ch <= '9':
        return ord(ch) - ord('0')
    elif 'A' <= ch <= 'F':
        return ord(ch) - ord('A') + 10
    elif 'a' <= ch <= 'f':
        return ord(ch) - ord('a') + 10
    else:
        raise ValueError(f"Invalid hexadecimal character: {ch}")


def hex_to_decimal(hex_number):
    if not isinstance(hex_number, str):
        raise ValueError("Input must be a string.")

    hex_number = hex_number.strip()

    if hex_number == "":
        raise ValueError("Empty input is not valid.")

    # Minus sign is allowed only at the beginning
    if "-" in hex_number[1:]:
        raise ValueError("Invalid format: '-' must appear only at the beginning.")

    # Optional sign handling
    sign = 1
    if hex_number[0] == "-":
        sign = -1
        hex_number = hex_number[1:]

    if hex_number == "":
        raise ValueError("Invalid input.")

    # Only one decimal point is allowed
    if hex_number.count(".") > 1:
        raise ValueError("Invalid hexadecimal format.")

    # Split into integer and fractional parts
    if "." in hex_number:
        integer_part, fractional_part = hex_number.split(".")
    else:
        integer_part = hex_number
        fractional_part = ""

    if integer_part == "":
        integer_part = "0"

    # Convert integer part
    decimal_value = 0
    for ch in integer_part:
        digit = hex_char_to_value(ch)
        decimal_value = decimal_value * 16 + digit

    # Convert fractional part
    fractional_value = 0
    power = 16

    for ch in fractional_part:
        digit = hex_char_to_value(ch)
        fractional_value += digit / power
        power *= 16

    return sign * (decimal_value + fractional_value)


def calculator():
    while True:
        print("\n=== Decimal-Hex Calculator ===")
        print("1. Decimal to Hexadecimal")
        print("2. Hexadecimal to Decimal")
        print("3. Exit")

        choice = input("Choose an option (1/2/3): ").strip()

        if choice == "1":
            user_input = input("Enter a decimal number: ").strip()
            try:
                if user_input == "":
                    raise ValueError("Empty input is not valid.")

                # Minus sign is allowed only at the beginning
                if "-" in user_input[1:]:
                    raise ValueError("Invalid format: '-' must appear only at the beginning.")

                # Only one decimal point is allowed
                if user_input.count(".") > 1:
                    raise ValueError("Invalid decimal format.")

                decimal_number = float(user_input)

                print("Hexadecimal:", decimal_to_hex(decimal_number))

            except ValueError as e:
                print("Error:", e)

        elif choice == "2":
            user_input = input("Enter a hexadecimal number: ").strip()
            try:
                print("Decimal:", hex_to_decimal(user_input))
            except ValueError as e:
                print("Error:", e)

        elif choice == "3":
            print("Exiting calculator.")
            break

        else:
            print("Error: Invalid choice.")

# Execution
calculator()
```
